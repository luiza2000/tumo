```python
import tensorflow as tf


input = tf.constant(3.2)
weight = tf.Variable(2.0)
output = tf.multiply(input,weight)



op = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(op)
    print(sess.run(output))

    output_ = tf.constant(10.0) 
    loss = (output - output_)**2
    optimizier = tf.train.GradientDescentOptimizer(0.025) 
    train_step = optimizier.minimize(loss) 
    for i in range(1,20):
        sess.run(train_step)
        print("Step"+str(i))
        print(sess.run(output))
        print(sess.run(weight))
```
